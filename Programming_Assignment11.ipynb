```python
#Question1 Write a Python program to find words which are greater than given length k?
str = "Shubham and surendra are good friends"
k = int(input())
strlist = str.split()
strlist1 = []
for i in strlist:
  if len(i)>k:
    strlist1.append(i)
print(' '.join(strlist1))
```

```python
#Question5 Write a Python program to find uncommon words from two Strings?
from collections import Counter
s1 = "Shubham and surendra are good friends"
s2 = "Shyam or surendra are not brother"
sa1 = s1.split()
sa2 = s2.split()
s3 = sa1+sa2
s4 = Counter(s3)
s5 = dict(s4)
lst = []
for k,v in s5.items():
    if v == 1:
        lst.append(k)
print(lst)

# x = [k for k,v in s5.items() if v == 1] #List Comphrension
# print(x)


 
```

```python
#Question6. Write a Python to find all duplicate characters in string?
from collections import Counter
s6 = "Shubham and surendra are good friends"
lst1=[]
s7 = dict(Counter(s6))
for k,v in s7.items():
  if v>1:
    lst1.append(k)
print(lst1)
```
